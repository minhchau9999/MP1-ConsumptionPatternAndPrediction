```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the CSV
df = pd.read_csv('electricity.csv', parse_dates=['timestamp'])
df.set_index('timestamp', inplace=True)
```

```python
df.head()
```

```python
# Transpose so that each row is a building
# Each row becomes a building, columns = hourly values
df_t = df.T
df_t.shape
```

```python
df_t.head()
```

```python
# Drop buildings with >20% missing values
df_t = df_t.dropna(thresh=df_t.shape[1] * 0.8)

# Fill small gaps
df_t = df_t.fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# normalize the load profiles


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_t)
```

#### Clustering using Kmeans

```python
# Elbow chart
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.xticks(K_range)
plt.show()
```

```python
# use optimal k = 4
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(df_scaled)

# Add labels back to the dataframe
df_t['cluster'] = labels
```

```python
import matplotlib.pyplot as plt

for i in range(4):
    cluster_data = df_t[df_t['cluster'] == i].drop('cluster', axis=1)
    mean_profile = cluster_data.mean()
    plt.plot(mean_profile.values, label=f'Cluster {i}')

plt.title("Average Load Profile by Cluster")
plt.xlabel("Hour Index")
plt.ylabel("kWh")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_t.head()
```

```python
from sklearn.decomposition import PCA

# Reduce to 2 dimensions
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)

# pca_result is now a (n_buildings x 2) matrix
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    pca_result[:, 0],
    pca_result[:, 1],
    c=labels,
    cmap='Set2',
    edgecolor='k',
    s=40,
    alpha=0.8
)

plt.title('PCA Projection of Building Load Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.colorbar(scatter, label='Cluster')
plt.tight_layout()
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_result = tsne.fit_transform(df_scaled)

plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=labels, cmap='Set2', alpha=0.7)
plt.title("t-SNE Projection of Load Clusters")
plt.grid(True)
plt.show()
```

##### Calendar heatmap of 1 building

```python
# 1. pick 1 building
# Get building IDs from one cluster
sample_building_id = df_t[df_t['cluster'] == 0].index[0]
print("Sample building:", sample_building_id)
```

```python
# 2. extract TS data for that building

df = pd.read_csv("electricity.csv", parse_dates=['timestamp'])
df.set_index('timestamp', inplace=True)

# Extract that building's load profile
ts = df[sample_building_id].copy()
```

```python
# 3. Format for calendar heatmap
calendar_df = ts.reset_index()
calendar_df.columns = ['timestamp', 'load']
calendar_df['date'] = calendar_df['timestamp'].dt.date
calendar_df['hour'] = calendar_df['timestamp'].dt.hour
calendar_df['dayofweek'] = calendar_df['timestamp'].dt.day_name()
calendar_df['month'] = calendar_df['timestamp'].dt.month_name()
calendar_df['year'] = calendar_df['timestamp'].dt.year
```

```python
# 4. aggregate by Day
# Option A: Total load per day
daily_load = calendar_df.groupby('date')['load'].sum().reset_index()

# Option B: Average load per hour-of-day across the year
hourly_avg = calendar_df.groupby(['date', 'hour'])['load'].mean().reset_index()
```
